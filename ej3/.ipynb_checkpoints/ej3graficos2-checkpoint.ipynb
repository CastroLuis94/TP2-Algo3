```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import math

tiemposcompleto = []
tiemposconexo = []
tiemposnoconexo = []


completo = pd.read_csv('ejercicio3_test2_completo.csv')
for i in range(2, 151):
    tam =  completo[completo["Ciudades"] == i]
    tiemposcompleto.append(tam["tiempo1"].mean())   

for i in range(2,149):
    tiemposcompleto[i] = tiemposcompleto[i]#/(i* math.log(i))
```

```python
conexo = pd.read_csv('ejercicio3_test2_conexo.csv')
for i in range(2, 151):
    tam =  conexo[conexo["Ciudades"] == i]
    tiemposconexo.append(tam["tiempo1"].mean())
for i in range(2,149):
    tiemposconexo[i] = tiemposconexo[i]#/(i* math.log(i))
```

```python
noconexo = pd.read_csv('ejercicio3_test2_no_conexo.csv')
for i in range(2, 151):
    tam =  noconexo[noconexo["Ciudades"] == i]
    tiemposnoconexo.append(tam["tiempo1"].mean())

for i in range(2,149):
    tiemposnoconexo[i] = tiemposnoconexo[i]#/(i*math.log(i))   
```
